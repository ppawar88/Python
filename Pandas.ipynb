```python
gold = pd.DataFrame({'Country': ['USA', 'France', 'Russia'],
                         'Medals': [15, 13, 9]}
                    )
silver = pd.DataFrame({'Country': ['USA', 'Germany', 'Russia'],
                        'Medals': [29, 20, 16]}
                    )
bronze = pd.DataFrame({'Country': ['France', 'USA', 'UK'],
                        'Medals': [40, 28, 27]}
                    )
```

```python
allMedals = gold.merge(silver, how="outer", on="Country")
allMedals = allMedals.merge(bronze, how="outer", on="Country")
allMedals.fillna(value=0, inplace=True)
allMedals["Total"] = allMedals["Medals_x"] + allMedals["Medals_y"] + allMedals["Medals"]
allMedals.sort_values(by="Total", ascending=False)
```

```python
resultDF = pd.DataFrame(data=allMedals, columns=["Country","Total"])
resultDF.set_index("Country", inplace=True)
resultDF.rename(columns={"Total":"Medals"}, inplace=True)
```
